```python
# import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# policies to compare

file1 = "regions.csv"
file2 = "regions_improved_2.csv" 
```

```python
# read the data

df1 = pd.read_csv(file1, delimiter=';')
df2 = pd.read_csv(file2, delimiter=';')
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
merged = pd.merge(df1, df2, on='id', suffixes=('_file1', '_file2'))
merged['charger_difference'] = merged['chargers_file2'] - merged['chargers_file1']

sum_charger_difference = merged['charger_difference'].sum()
print("The total number of chargers added is:", sum_charger_difference)
```

```python
# plot the chargers added per region

plt.figure(figsize=(8, 4))
plt.bar(merged['id'], merged['charger_difference'])
plt.ylabel('Chargers')
plt.title('Charger Additions per Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
