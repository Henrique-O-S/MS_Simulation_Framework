```python
# import libraries

import json
```

```python
# load roads and journeys data

with open("roads.json") as f:
    roads = json.load(f)

with open("journeys.json") as f:
    journeys = json.load(f)
```

```python
# calculate score for each journey, based on path traffic and path length

length_penalizer = 2

def calculate_traffic_score(path, roads):
    total_traffic = 0
    for i in range(len(path) - 1):
        start, end = path[i], path[i + 1]
        total_traffic += roads[start][end]  
    path_length = len(path) - 1  
    return round(total_traffic / (path_length ** length_penalizer), 2)
```

```python
# for each region, calculate the probabilities of each journey based on the scores

def calculate_probabilities(journeys, roads):
    for region, data in journeys.items():
        paths = data["paths"]
        scores = [calculate_traffic_score(path, roads) for path in paths]
        total_score = sum(scores)
        probabilities = [round(score / total_score, 2) for score in scores] if total_score > 0 else []
        journeys[region]["probabilities"] = probabilities
        print('-' * 50)
        print(region.capitalize(), '\n')
        for path, score in zip(paths, scores):
            print(path, score)
        print("\nProbabilities:", probabilities)
```

```python
# probabilities calculation 

calculate_probabilities(journeys, roads)
```
